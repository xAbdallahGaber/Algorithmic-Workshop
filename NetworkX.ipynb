```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from(['A','B','C','D','E','F'])
G.add_edges_from([('A', 'C',{'weight': 4}), ('A', 'D',{'weight': 4}), ('C', 'D',{'weight': 2}), ('B', 'D',{'weight': 3})
                 , ('B', 'E',{'weight': 3}), ('E', 'D',{'weight': 4}), ('F', 'D',{'weight': 2}), ('F', 'E',{'weight': 3})])
pos=nx.spring_layout(G)
labels = nx.get_edge_attributes(G,'weight')

plt.figure(1)
nx.draw(G,pos,with_labels=True)
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)

dijkstraFrom = 'A'
dijkstraTo = 'B'


print("DFS output:",list(nx.dfs_successors(G,'A')))
print("BFS output:",list(nx.bfs_successors(G,'A')))
print("Dijkstra's path from {} to {}:".format(dijkstraFrom,dijkstraTo),list(nx.dijkstra_path(G,dijkstraFrom,dijkstraTo)))

plt.figure(2)
primSubEdges = list(nx.minimum_spanning_edges(G, algorithm="prim", data=False))
print('Prim Edges (fig2): ',primSubEdges)
primG = nx.Graph()
primG.add_nodes_from(G.nodes)
primG.add_edges_from(primSubEdges)
nx.draw(primG,pos,with_labels=True)

plt.figure(3)
kruskalSubEdges = list(nx.minimum_spanning_edges(G, algorithm="kruskal", data=False))
print('krsukal Edges (fig3): ',kruskalSubEdges)
kruskalG = nx.Graph()
kruskalG.add_nodes_from(G.nodes)
kruskalG.add_edges_from(kruskalSubEdges)
nx.draw(kruskalG,pos,with_labels=True)

```
