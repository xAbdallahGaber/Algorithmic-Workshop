```python
def Solve(myZip,weights,maxWeight,isFractional): 
    ''' 
    myZip = zip(index,loss,itemWeight,itemProfit) 
    Author: Abdallah Gaber
    '''
    print('*'*100)
    addedIteams = []
    currentWeight = 0
    currentProfit = 0
    sortedIteams = sorted(list(myZip),key = lambda i: i[1],reverse = True)
    for item in sortedIteams:
        if currentWeight+item[2]<=maxWeight:
            addedIteams.append(item[0])
            currentWeight+= item[2]
            currentProfit+= item[3]
        elif isFractional:
            partTaken = (maxWeight-currentWeight)/item[2]
            addedIteams.append(item[0])
            currentWeight+= item[2]*partTaken
            currentProfit+= item[3]*partTaken
            print('Last iteam is taken by a part of',partTaken)
            break
    print('Carried Weight:',currentWeight)
    print('Carried Profit:',currentProfit)
    print('Carried Items:',addedIteams)
    print('SOLVED!')
    print('*'*100)

def inputIsFractional():
    while True:
        userIn = input("Enter (0) for Factorial, (1) for I/0:")
        if userIn == '0' or userIn == '1':
            if userIn == '0':
                return True
            else:
                return False
        else:
            print("INVALID INPUT! Try again.")
def inputCostFx():      
    while True:
        userIn = input("Enter (0) for Maximum Profit, (1) for Minimum Weight:, (2) for Maximum Profit/Weight:")
        if userIn == '0' or userIn == '1' or userIn == '2':
            return int(userIn)
        else:
            print("INVALID INPUT! Try again.")
    
def inputConstrain():
    itemWeight = []
    itemProfit = []
    maxWeight = int(input("Enter Maximum Weight:"))
    numOfItems = int(input("Enter number of items:"))
    for i in range(numOfItems):
        itemWeight.append(int(input("Item# {} : Enter Weight:".format(i))))
        itemProfit.append(int(input("Item# {} : Enter Profit:".format(i))))

    return (maxWeight,itemProfit,itemWeight)

def KnapsackDynamic(itemProfit,itemWeight,maxWeight):
    m = [[0 for col in range(maxWeight+1)] for row in itemProfit+['']]
    for i in range(1,len(m)):
        for j in range(1,len(m[0])):
            if(itemWeight[i-1]>j):
                m[i][j] = m[i-1][j]
            else:
                m[i][j] = max(m[i-1][j], itemProfit[i-1]+m[i-1][j-itemWeight[i-1]])
    return(m[len(itemWeight)][maxWeight])

def Knapsack():
    #Editable Constrains
    isFractional = True
    costFx = 0 
    maxWeight = 35
    itemProfit = [5,8,3,2,7,9,4]
    itemWeight = [7,8,4,10,4,6,4]
    itemIndex = [i for i in range(len(itemProfit))]
    
    itemWeightInv = [1/i for i in itemWeight]
    itemProfitWeight = [a/b for a,b in zip(itemProfit, itemWeight)]
    
    print("Welcome to knapsack solution using greedy algorithm: ")
    print("Default parameters:")
    print("-Cost fx is Maximum Profit")
    print("-Maximum Weight",maxWeight)
    print("-items profit", itemProfit)
    print("-items weight", itemWeight)
    
    while True:
        print('#'*100)
        print("Please Select an option: ")
        print('0: Solve')
        print('1: Change the type of Knapsack (Fractional - I/0)')
        print('2: Change the cost function')
        print('3: Update the objects and the maximum weight of the knapsack')
        print('4: Close the program')
        print('5: Solve Dynamic Progrraming')
        userIn = input("Enter your selection: ")
        
        if userIn == '0':
            if costFx == 0: #Max Profit
                myZip = zip(itemIndex,itemProfit,itemWeight,itemProfit)
            elif costFx == 1: #Min Weight
                myZip = zip(itemIndex,itemWeightInv,itemWeight,itemProfit)
            elif costFx == 2: #Max Profit/Weight
                myZip = zip(itemIndex,itemProfitWeight,itemWeight,itemProfit)
            Solve(myZip,itemWeight,maxWeight,isFractional)
            
        elif userIn == '1':
            isFractional = inputIsFractional()
            print("UPDATED!")
            continue
            
        elif userIn == '2': 
            costFx = inputCostFx()
            print("UPDATED!")
            continue
            
        elif userIn == '3': 
            (maxWeight,itemProfit,itemWeight) = inputConstrain()
            itemIndex = [i for i in range(len(itemProfit))]
            itemWeightInv = [1/i for i in itemWeight]
            itemProfitWeight = [a/b for a,b in zip(itemProfit, itemWeight)]
            print("UPDATED!")
            continue
            
        elif userIn == '4': 
            print("CLOSING...")
            return
        
        elif userIn == '5': 
            print("SOLVED with Dynamic Programming! carried profit: ",KnapsackDynamic(itemProfit,itemWeight,maxWeight))
             
        else:
            print('INVALID INPUT. Try Again.')
            continue
            
            
Knapsack()
```
